```python
import numpy as np
import math

# src https://math.stackexchange.com/questions/995659/given-two-points-find-another-point-a-perpendicular-distance-away-from-the-midp
def perpendicular(x1,y1,x2,y2,x3,y3, d):
    k = d / math.sqrt((y1-y2)**2 + (x2-x1)**2)
    x = x3 + k*(y1-y2)
    y = y3 + k*(x2-x1)
    return [x,y]
```

```python
df_track = pd.DataFrame(columns=['left_x','left_y', 'right_x', 'right_y'])
df = df_read

for indx in df_read.index:
    if indx == 0: 
        indx = 1
    perp_l = perpendicular( df["# x_m"][indx-1], df["y_m"][indx-1] , df["# x_m"][indx+1], df["y_m"][indx+1], df["# x_m"][indx], df["y_m"][indx], df["w_tr_left_m"][indx]*4)
    perp_r = perpendicular( df["# x_m"][indx-1], df["y_m"][indx-1] , df["# x_m"][indx+1], df["y_m"][indx+1], df["# x_m"][indx], df["y_m"][indx], -df["w_tr_right_m"][indx]*4)
    df_track = df_track.append({'left_x': perp_l[0], 'left_y': perp_l[1], 'right_x': perp_r[0], 'right_y': perp_r[1], 'x_m': df["# x_m"][indx], 'y_m': df["y_m"][indx],}, ignore_index=True)

```

**Build the autoencoder**

```python
import keras
from keras import layers
from keras.models import Model, Sequential
from keras.layers import Dense, Conv2D, Dropout, BatchNormalization, Input, Reshape, Flatten, Conv2DTranspose, MaxPooling2D, UpSampling2D
from keras.optimizers import Adam
import tensorflow

from sklearn.model_selection import train_test_split
```

```python
#ENCODER
# inp = Input((450, 450,1))
input_img = keras.Input(shape=(144, 144, 3))

x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(input_img)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(x)

x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(x)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = layers.MaxPooling2D((2, 2), padding='same')(x)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(x)

x = layers.MaxPooling2D((2, 2), padding='same')(x)
x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = layers.UpSampling2D((2, 2))(x)


x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(x)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = layers.UpSampling2D((2, 2))(x)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(x)

x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(x)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)

decoded = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

autoencoder = keras.Model(input_img, decoded)
# binary_crossentropy
opt = keras.optimizers.Adam()
autoencoder.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])

autoencoder.summary()
```

```python
import glob
x = sorted(glob.glob('./thick/sv/x/*.*'))
y = sorted(glob.glob('./thick/sv/y/*.*'))

data = []
labels = []
step = 1

for i in range(0, len(x), step):
    image=tensorflow.keras.preprocessing.image.load_img(x[i], 
    target_size= (144,144,3))
    image=np.array(image)/255
    data.append(image)
    
    print(f"{x[i]} vs {y[i]} indx={i}")
    image_y=tensorflow.keras.preprocessing.image.load_img(y[i], 
    target_size= (144,144,3))
    image_y=np.array(image_y)/255
    labels.append(image_y)

    
sx = sorted(glob.glob('./thick/mz/x/*.*'))
sy = sorted(glob.glob('./thick/mz/y/*.*'))

for i in range(0, len(sx), step):   
    image=tensorflow.keras.preprocessing.image.load_img(sx[i], 
    target_size= (144,144,3))
    image=np.array(image)/255
    data.append(image)
    
    print(f"{sx[i]} vs {sy[i]} indx={i}")
    image_y=tensorflow.keras.preprocessing.image.load_img(sy[i],
    target_size= (144,144,3))
    image_y=np.array(image_y)/255
    labels.append(image_y)

data = np.array(data)
labels = np.array(labels)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(data, labels, test_size=0.33, shuffle= True)
```
